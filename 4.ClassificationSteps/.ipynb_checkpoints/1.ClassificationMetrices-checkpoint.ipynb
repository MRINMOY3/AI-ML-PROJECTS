### Evaluation of Classification Project

### Accuracy Score

$$\text {Accuracy Score } = \frac { \text {correct predictions} } { \text {total observation}}  $$

```python
def accuracy_score(y, y_hat):
    return (y == y_hat).sum() / len(y)
```

```python
print(f"Train Accuracy: {accuracy_score(y_train, y_hat_train):.2f}")
print(f"Test Accuracy: {accuracy_score(y_test, y_hat_test):.2f}")
```

```python
X = iris.drop('species', axis=1)
y = iris['species']
```

```python
from sklearn.model_selection import StratifiedShuffleSplit

def train_test_split(X, y):
    split = StratifiedShuffleSplit(n_splits=1, random_state=123, test_size=0.2)
    (tr_idx, ts_idx), = split.split(X, y)
    X_train = X.iloc[tr_idx]
    y_train = y.iloc[tr_idx]
    X_test  = X.iloc[ts_idx]
    y_test  = y.iloc[ts_idx]
    return X_train, X_test, y_train, y_test

```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y)
```

```python
def plot_pr_curve(pr, re, th):
    plt.figure(dpi=100)
    plt.plot(th, pr[:-1], 'r--', label='precision')
    plt.plot(th, re[:-1], 'g:', label='recall')
    plt.vlines(0.5, 0, 1, ls='--', color='k')
    plt.legend()
    plt.grid()
    plt.axis([0, 1.25, 0, 1.25])
    plt.xticks(np.arange(0, 1.1, .1))
    plt.yticks(np.arange(0, 1.1, .1))
    plt.tight_layout()
    plt.show()
```

```python
precisions, recalls, thresholds = precision_recall_curve(y_train, true)
plot_pr_curve(precisions, recalls, thresholds)
```

### PR CURVE

```python
plt.plot(recalls, precisions, label='PR Curve')
plt.xlabel("Recall")
plt.ylabel("Preicsion")
plt.grid()
plt.show()
```
