# LASSO Regression

### Least Absolute Shrinkage Selection Operator Regression

#### Ridge 

$J\theta = MSE \theta + \alpha \sum\limits_{i=0}^N {\theta_i}^2 $

#### Lasso Regression 
    l1 penalty, auto feature selection 
$J\theta = MSE \theta + \alpha \sum\limits_{i=0}^N |\theta_i| $

#### infographic  / cheetsheet

```python
from sklearn.linear_model import Lasso, Ridge
from sklearn.linear_model import SGDRegressor
```

```python
ridge = Ridge(alpha=0.1)
lasso = Lasso(alpha=0.2)

# hyper tune alpha values

sgd_ridge = SGDRegressor(penalty='l2', alpha=0.1)
sgd_lasso = SGDRegressor(penalty='l1', alpha=0.1)
```

    Lasso use to drop features as it fits a regression line sometime it can drop features which can play a big role
    
    Lasso bounce around your minimum point so you need to use it with early stopping techniques
    
    When features are greater than observations Lasso perform very poor

### Elastic Net 


    it is a mixture of Ridge and Lasso regression 

$J\theta = MSE \theta + \alpha \sum\limits_{i=0}^n {\theta}^2 + \alpha \sum \limits_{i=0}^n |\theta|$

```python
from sklearn.linear_model import ElasticNet
from sklearn.linear_model import SGDRegressor
```

```python
elastic_ridge = ElasticNet(alpha=0.1, l1_ratio=0) # Ridge
elastic_lasso = ElasticNet(alpha=0.1, l1_ratio=1) # Lasso

elastic_model = ElasticNet(alpha=0.1, l1_ratio=0.5) # equally Ridge Equally Lasso
```

    0.1 -> less LASSO, High Ridge
    
    0.7 -> High LASSO, Low Ridge

```python
elastic_sgd_model = SGDRegressor(alpha=0.1, l1_ratio=0.15, penalty='elasticnet')
```

```python
import numpy as np
```

```python
hours = np.concatenate((np.random.normal(3, 1.5, 50), np.random.normal(10, 2, 50))).reshape(-1, 1)
result = np.concatenate((np.zeros(50), np.ones(50)))
```

```python
plt.figure(dpi=100)
sns.regplot(hours, result, line_kws={'color':'red'})
plt.xlabel("Study Hours")
plt.ylabel("Result")
plt.yticks([0, 1], ['Fail', 'Pass'])
plt.title("Study Hours vs Result")
plt.grid()
plt.show()
```

```python
from sklearn.linear_model import SGDRegressor
```

```python
model = SGDRegressor()
model.fit(hours, result)
```

```python
y_hat = model.predict(hours)
```

```python
plt.figure(dpi=100)
plt.scatter(hours, result, marker='x')
plt.plot(hours, y_hat, 'r--')
plt.grid()
plt.show()
```

```python
def predict(y_hat, th=0.51):
    result = []
    for value in y_hat:
        if value >= th:
            result.append(1.0)
        else:
            result.append(0.0)
    return result
```

```python
labels = predict(y_hat)
```

```python
labels[:5]
```

```python
result[:5]
```

```python
acc = (labels == result).sum() / len(result)
```

```python
print(acc)
```

### Sigmoid

 $$F(z) = \frac {1} {1+e-{z}}$$

$ H(\theta) = \theta_0 + \theta_1 X1$


$ \sigma = \frac {1} {1+e^{-z}}$

$\hat p = \sigma (H\theta) $

$$\hat p = \frac {1} {1+e^{-(\theta_0 + \theta_1 X_1)}}$$

0<=sigmoid(z)<=1

returns an s shape curve

```python
arr = np.linspace(-5, 5, 100)
```

```python
def sigmoid(z):
    return 1 / (1+np.exp(-z))
```

```python
rs = sigmoid(arr)
```

```python
plt.figure(dpi=100)
plt.plot(arr)
plt.show()
```

```python
plt.figure(dpi=100)
plt.title("Sigmoid")
plt.plot(rs)
plt.show()
```

```python
y_pred = sigmoid(y_hat)
```

    Log Error

```python
plt.figure(dpi=100)
plt.scatter(hours, result, marker='x')
plt.plot(hours, y_hat, 'r--')
plt.scatter(hours, y_pred, marker='o')
plt.grid()
plt.show() # Mean Squared Error
```
