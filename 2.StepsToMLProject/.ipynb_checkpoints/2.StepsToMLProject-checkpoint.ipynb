### Regression

    predicting value of a continous dependent variable on the basis of independent variables 

```python
height = np.random.normal(170, 15, 100)

bias = np.random.random(100) * 10
weight = 0.6 * height + bias

corr_plot(height, weight, 'Height', 'Weight')
```

```python
#bias.mean()
```

```python
def MAE(y, y_hat):
    error = (np.abs(y-y_hat)).sum() / len(y)
    return error
```

```python
def predict(m, c, X):
    y_hat = m * X + c
    return y_hat
```

```python
def corr_plot(X, y, xlb='', ylb=''):
    r = np.corrcoef(X, y)[0, 1]
    plt.figure(dpi=100)
    plt.scatter(X, y, label=f'r={r:.2f}')
    plt.xlabel(xlb)
    plt.ylabel(ylb)
    return plt.gca()
```

#### Linear Regression

    if your feature and target has a linear releationship than you can predict value of target on the basis of feature
    
    to measure linear dependency we can calculate Correlation or Covariance 

```python
ax = corr_plot(height, weight)

m = 0.2; c = 2
y_hat = predict(m, c, height)
error = MAE(weight, y_hat)
plt.plot(height, y_hat, 'c--', label=f'error = {error:.2f}')

m = 0.4; c = 3 # parameter
y_hat = predict(m, c, height)
error = MAE(weight, y_hat)
plt.plot(height, y_hat, 'g--', label=f'error = {error:.2f}')

m = 0.5; c = 4
y_hat = predict(m, c, height)
error = MAE(weight, y_hat)
plt.plot(height, y_hat, 'b--', label=f'error = {error:.2f}')


m = 0.6; c = 4.7 # Machine Learning, Parameters
# weight= m * height + c
y_hat = predict(m, c, height)
error = MAE(weight, y_hat)
plt.plot(height, y_hat, 'r-', label=f'error = {error:.2f}')

plt.legend()
plt.grid()
plt.show()
```

### Prediction Function / Hypothesis Function

$ \hat y = H_{\theta}(X) = \theta_0 + \theta_1 X_1 $


### Cost Functions 

    we estimate parameter values by minimizing cost function, so cost function for a algorithm should be diffrentiable
    
    if you select convex function as cost function that will be best 
    
    
    
$ J(\theta) = MSE(\theta) = \frac {\sum \limits_{i=0}^N (\hat y_i - y_i)} {N} $

$ J(\theta) = MSE(\theta) = \frac {\sum \limits_{i=0}^N (\theta_0 + \theta_1 X_{1_{i}} - y_i)} {N} $

### We have a direct solutions to this problem 

# Normal Equation

$\theta_1 = \frac {\sum\limits_{i=0}^N (\bar y - y_i) (\bar X_1 - X_{1_i}) } { {\sum\limits_{i=0}^N(\bar X_1 - X_{1_i})}^2 } $

$ \theta_0 = \bar y - \theta_0 \bar X_1 $

### fitting

    estimating value of parameters using training data is known as learning or fitting

    90 % -> Product Based 
    
    10 % -> Research and Development 

```python
class LinearRegression:
    def __init__(self):
        self.theta_0 = None
        self.theta_1 = None
    def fit(self, X, y):
        """estimate theta values"""
        xbar = X.mean()
        ybar = y.mean()
        
        self.theta_1 = ( (ybar-y)*(xbar-X)).sum() / ( (xbar-X)**2 ).sum()
        self.theta_0 = ybar - self.theta_1 * xbar
        
    def predict(self, X):
        y_hat = self.theta_0 + self.theta_1 * X
        return y_hat
    
    
```

```python
X = height
y = weight
```

```python
lin_model = LinearRegression()
```

#### fitting


```python
lin_model.fit(X, y)
```

```python
lin_model.theta_0
```

```python
lin_model.theta_1
```

```python
y_hat = lin_model.predict(X)
```
